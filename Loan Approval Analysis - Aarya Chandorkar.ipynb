# HANDLING MISSING DATA

```python
# Categorical Columns - Using Mode
df['Gender'].fillna(df.Gender.mode()[0], inplace=True)  # Fills missing Gender with the most frequent value
df['Married'].fillna(df.Married.mode()[0], inplace=True)  # Fills missing Married status with the most frequent value
df['Dependents'].fillna(df.Dependents.mode()[0], inplace=True)  # Fills missing Dependents count with the most frequent value
df['Self_Employed'].fillna("Unknown", inplace=True)  # Fills missing Self_Employed status with 'Unknown'  
```

```python
# Numerical Columns - Using Median
df['LoanAmount'].fillna(df.LoanAmount.median(), inplace=True)  # Fills missing LoanAmount with the median value
df['Loan_Amount_Term'].fillna(df.Loan_Amount_Term.mode()[0], inplace=True)  # Fills missing Loan Amount Term with the most frequent value
df['Credit_History'].fillna(df.Credit_History.mode()[0], inplace=True)  # Fills missing Credit History with the most frequent value
```

```python
# Check if all missing values are filled
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df["CoapplicantIncome"] = df["CoapplicantIncome"].astype("int64")
df["LoanAmount"] = df["LoanAmount"].astype("int64")
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].astype("int64")
df['Credit_History'] = df['Credit_History'].astype(int)
```

```python
df.dtypes
```

```python
# Replacing categorical values with more meaningful labels for better clarity
df["Loan_Approval"] = df["Loan_Status"].replace({"Y": "Approved", "N": "Not Approved"})
df["Gender"] = df["Gender"].replace({"Male": "M", "Female": "F"})
df["Married"] = df["Married"].replace({"Married": "Married", "Not Married": "Single"})
df["Dependents"] = df["Dependents"].replace({"0": "No Dependents", "1": "1 Dependent", "2": "2 Dependents", "3+": "3 or more Dependents"})
df["Education"] = df["Education"].replace({"Graduate": "Graduated", "Not Graduate": "Non-Graduated"})
df["Self_Employed"] = df["Self_Employed"].replace({"Yes": "Employed", "No": "Not Employed"})
df["Property_Area"] = df["Property_Area"].replace({"Urban": "Urban Area", "Semiurban": "Semiurban Area", "Rural": "Rural Area"})
```

```python
# Examine the influence of the number of dependents on loan approval
plt.figure(figsize=(8, 6))
sb.countplot(data=df, x="Dependents", hue="Loan_Status", palette="Pastel1")
plt.title("Loan Approval by Number of Dependents")
plt.xlabel("Number of Dependents")
plt.ylabel("Count")
plt.show()
```

```python
# Investigate loan approval by education background
plt.figure(figsize=(8, 6))
sb.countplot(data=df, x="Education", hue="Loan_Status", palette="Set2")
plt.title("Loan Approval by Education Background")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

```python
# Assess the impact of self-employment status on loan approval
plt.figure(figsize=(8, 6))
sb.countplot(data=df, x="Self_Employed", hue="Loan_Status", palette="coolwarm")
plt.title("Loan Approval by Self-Employment Status")
plt.xlabel("Self-Employed")
plt.ylabel("Count")
plt.show()
```

# Income and Loan Amount Analysis

```python
# Visualize the relationship between applicant income and loan approval using boxplots
plt.figure(figsize=(8, 6))
sb.boxplot(data=df, x="Loan_Status", y="ApplicantIncome", palette="Blues")
plt.title("Loan Approval vs Applicant Income")
plt.xlabel("Loan Status")
plt.ylabel("Applicant Income")
plt.show()
```

```python
# Examine how co-applicant income influences loan approval
plt.figure(figsize=(8, 6))
sb.boxplot(data=df, x="Loan_Status", y="CoapplicantIncome", palette="Greens")
plt.title("Loan Approval vs Coapplicant Income")
plt.xlabel("Loan Status")
plt.ylabel("Coapplicant Income")
plt.show()
```

```python
# Perform correlation analysis between applicant income, co-applicant income, and loan amount
correlation_matrix = df[["ApplicantIncome", "CoapplicantIncome", "LoanAmount"]].corr()
plt.figure(figsize=(8, 6))
sb.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Applicant Income, Coapplicant Income, and Loan Amount")
plt.show()
```

```python
# Compare loan amounts requested by different demographic groups (gender, marital status, education)
# Groupby Gender and analyze loan amount
gender_loan_amount = df.groupby('Gender')['LoanAmount'].mean()
gender_loan_amount.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'lightpink'])
plt.title("Average Loan Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Loan Amount")
plt.show()
```

```python
# Groupby Marital Status and analyze loan amount
marital_loan_amount = df.groupby('Married')['LoanAmount'].mean()
marital_loan_amount.plot(kind='bar', figsize=(8, 6), color=['lightgreen', 'salmon'])
plt.title("Average Loan Amount by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Average Loan Amount")
plt.show()
```

```python
# Groupby Education and analyze loan amount
education_loan_amount = df.groupby('Education')['LoanAmount'].mean()
education_loan_amount.plot(kind='bar', figsize=(8, 6), color=['lightyellow', 'lightcoral'])
plt.title("Average Loan Amount by Education Background")
plt.xlabel("Education")
plt.ylabel("Average Loan Amount")
plt.show()
```
